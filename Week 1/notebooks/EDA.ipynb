##### Trend analysis over Time
Daily, weekly, monthly and yearly

Steps
* Aggregate daily, weekly, monthly and yearly data
* Visualize the aggregated trend over time

##### Daily Aggregate

```python
# Ensure the index is datetimeidex for sampling by date or month
data.index = pd.to_datetime(data.index, errors='coerce',format='%Y-%m-%d %H:%M:%S%z')

# Aggregate data by date (D)
daily_counts = data.resample('D').size().reset_index(name='Number of Publications')
print(daily_counts)
# daily_counts.set_index('date', inplace=True)

# import plotly.io as pio
# pio.renderers.default = 'plotly_mimetype'

# Visualize the trends with Plotly
fig = px.line(daily_counts, x='date', y='Number of Publications', 
        title = 'Number of Publications over Time',
        labels = {'date':'Date', 'Number of Publications': 'Number of Publications'},
        template = 'seaborn' # plotly_dark, white, ggplot2, seaborn, simple_white 
        )
fig.update_layout(
    xaxis_title = 'Date', 
    yaxis_title='Number of Publications', 
    xaxis_tickangle=-45, # Rotate x-axis labels for better readability
    xaxis=dict(
        tickformat='%Y-%m-%d', # Ensure date is formated correctly
        dtick="M1", # Show a tick mark for each month
        tickvals = daily_counts['date'][::int(len(daily_counts)/10)].tolist() # Show fewer ticks
        ), 
    yaxis = dict(
        rangemode='tozero' # Ensure the y-axis starts from zero
    ),
    autosize=True
    )
fig.show()
        
```
