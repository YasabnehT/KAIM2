#### Timeseries Analysis

```python
data_ben_clean.set_index('Timestamp',inplace=True)
```

##### Line Plots

```python
# Plotting GHI, DNI, DHI, and Tamb over time
fig = go.Figure()

# Adding GHI, DNI, DHI to the plot
fig.add_trace(go.Scatter(x=data_ben_clean.index, y=data_ben_clean['GHI'], mode='lines', name='GHI', line=dict(color='orange')))
fig.add_trace(go.Scatter(x=data_ben_clean.index, y=data_ben_clean['DNI'], mode='lines', name='DNI', line=dict(color='red')))
fig.add_trace(go.Scatter(x=data_ben_clean.index, y=data_ben_clean['DHI'], mode='lines', name='DHI', line=dict(color='blue')))

# Adding Tamb to a secondary y-axis
fig.update_layout(
    title="Solar Irradiance Components (GHI, DNI, DHI) and Ambient Temperature Over Time",
    xaxis_title="Time",
    yaxis_title="Irradiance (W/m^2)",
    yaxis2=dict(title="Temperature (°C)", overlaying='y', side='right'),
    legend=dict(x=0.01, y=0.99),
)
# Adding Tamb to the plot using secondary y-axis
fig.add_trace(go.Scatter(x=data_ben_clean.index, y=data_ben_clean['Tamb'], mode='lines', name='Tamb', yaxis="y2", line=dict(color='green')))

fig.show()
```

##### Area Plots

```python
# Plotting GHI, DNI, DHI, and Tamb over time
fig = go.Figure()

# Adding GHI, DNI, DHI to the plot
fig.add_trace(go.Scatter(x=data_ben_clean.index, y=data_ben_clean['GHI'], mode='none', fill='tozeroy', name='GHI', line=dict(color='orange')))
fig.add_trace(go.Scatter(x=data_ben_clean.index, y=data_ben_clean['DNI'], mode='none', fill='tozeroy', name='DNI', line=dict(color='red')))
fig.add_trace(go.Scatter(x=data_ben_clean.index, y=data_ben_clean['DHI'], mode='none', fill='tozeroy', name='DHI', line=dict(color='blue')))

# Adding Tamb to a secondary y-axis
fig.update_layout(
    title="Solar Irradiance Components (GHI, DNI, DHI) and Ambient Temperature Over Time",
    xaxis_title="Time",
    yaxis_title="Irradiance (W/m^2)",
    yaxis2=dict(title="Temperature (°C)", overlaying='y', side='right'),
    legend=dict(x=0.01, y=0.99),
)
# Adding Tamb to the plot using secondary y-axis
fig.add_trace(go.Scatter(x=data_ben_clean.index, y=data_ben_clean['Tamb'], mode='lines', name='Tamb', yaxis="y2", line=dict(color='green')))

fig.show()
```

#### Correlation Analysis

##### Heatmap

```python
# Step 1: Correlation Heatmap using Plotly

# Selecting relevant columns for the heatmap
corr_data = data_ben_clean[['GHI', 'DNI', 'DHI', 'TModA', 'TModB']]

# Calculating the correlation matrix
corr_matrix = corr_data.corr()

# Plotly Heatmap
heatmap = go.Figure(data=go.Heatmap(
                   z=corr_matrix.values,
                   x=corr_matrix.columns,
                   y=corr_matrix.columns,
                   colorscale='Viridis'))

heatmap.update_layout(title='Correlation Heatmap between Solar Radiation Components and Temperature Measures')
heatmap.show()
```

##### Pair Plot

```python
# Step 2: Pairplot using Plotly

# Selecting relevant columns for the pairplot
pairplot_data = data_ben_clean[['GHI', 'DNI', 'DHI', 'WS', 'WSgust', 'WD', 'TModA', 'TModB']]

# Creating a pairplot (scatter matrix) using Plotly Express
pairplot = px.scatter_matrix(pairplot_data, dimensions=pairplot_data.columns, title="Pairplot: Solar Radiation, Temperature Measures, and Wind Conditions")

# Updating layout for better visuals
pairplot.update_traces(diagonal_visible=False)  # Hide diagonal plots for better clarity
pairplot.update_layout(width=1200, height=1200)
pairplot.show()
```

##### Temperature Analysis

##### Heatmap

```python
# Correlation Analysis between RH, Temperature, and Solar Radiation
corr_data = data_ben_clean[['RH', 'GHI', 'DNI', 'DHI', 'TModA', 'TModB']]
corr_matrix = corr_data.corr()

# Plotly Heatmap for Correlation Matrix
heatmap = go.Figure(data=go.Heatmap(
    z=corr_matrix.values,
    x=corr_matrix.columns,
    y=corr_matrix.columns,
    colorscale='Viridis'))

heatmap.update_layout(title='Correlation Heatmap: Influence of RH on Temperature and Solar Radiation')
heatmap.show()
```

##### Pair Plot

```python
# Scatter Plots for RH vs Temperature and RH vs Solar Radiation
scatter_fig = px.scatter_matrix(data_ben_clean, 
                                dimensions=['RH', 'GHI', 'DNI', 'DHI', 'TModA', 'TModB'], 
                                title="Scatter Matrix: Influence of RH on Temperature and Solar Radiation")

scatter_fig.update_traces(diagonal_visible=False)
scatter_fig.update_layout(width=1200, height=1200)
scatter_fig.show()
```

##### RH Influence - Heatmap

```python
# Heatmap for RH vs Temperature (TModA)
heatmap_tmoda = go.Figure(data=go.Histogram2d(
    x=data_ben_clean['RH'], 
    y=data_ben_clean['TModA'], 
    colorscale='Viridis'))

heatmap_tmoda.update_layout(
    title='Heatmap: Relative Humidity (RH) vs Temperature (TModA)',
    xaxis_title='Relative Humidity (%)',
    yaxis_title='Temperature of Module A (°C)',
)
heatmap_tmoda.show()

# Heatmap for RH vs Solar Radiation (GHI)
heatmap_ghi = go.Figure(data=go.Histogram2d(
    x=data_ben_clean['RH'], 
    y=data_ben_clean['GHI'], 
    colorscale='Viridis'))

heatmap_ghi.update_layout(
    title='Heatmap: Relative Humidity (RH) vs Global Horizontal Irradiance (GHI)',
    xaxis_title='Relative Humidity (%)',
    yaxis_title='Global Horizontal Irradiance (W/m^2)',
)
heatmap_ghi.show()
```
